# First Page Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
column_names = ['Title', 'Link', 'Price', 'Rating', 'Stock']

books_information = []

url = 'http://books.toscrape.com/catalogue/page-1.html'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

books = soup.findAll('article', {'class':'product_pod'})

for i in books:
        title = i.h3.a['title']
        # print(title)

        link = 'https://books.toscrape.com/catalogue/' + i.h3.a['href']
        # print(link)

        price = i.find('p',{'class':'price_color'}).text[2:]
        # print(price)

        rating = i.find('p', attrs={"class": "star-rating"})['class'][1]
        all_rating = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
        numRating = all_rating[rating]
        # print(numRating)
        
        stock = i.find('p',{'class':'instock availability'}).text.strip()
        # print(stock)

        books_information.append([title, link, price,numRating, stock])
        # print(title, link, price,numRating, stock)
  
```

```python
df = pd.DataFrame(books_information, columns = column_names)
```

```python
df.to_csv('Books Information.csv', index = False)
```
